```python
import pandas as pd
```

```python
df = pd.read_csv("./Data Tidying Examples - Sidewinder Library.csv")
```

```python
df.info()
```

```python
library_patrons_df = df[["Last Name", "First Name", "Date of Birth", "Library Card Expired?"]].copy().drop_duplicates()
# library_patrons_df["Full Name"] = df["First Name"] + " " + df["Last Name"]
# library_patrons_df.sort_values("Full Name")
library_patrons_df.sort_values(by=["Last Name", "First Name"])
```

```python
library_patrons_df.to_csv("library_patrons.csv")
```

```python
library_books_df = df[["Genres", "Book", "Author_A", "Author_B"]].copy().drop_duplicates()
library_books_df.sort_values(["Genres", "Book", "Author_A"])
```

```python
library_books_df.to_csv("library_books.csv")
```

```python
library_fees_df = df[["Last Name", "First Name", "Book", "Days Late", "Late Fees"]].copy()
# library_fees_df["Full Name"] = df["First Name"] + " " + df["Last Name"]
library_fees_df.sort_values(["Last Name", "First Name", "Book"])
```

```python
library_fees_df.to_csv("library_fees.csv")
```
